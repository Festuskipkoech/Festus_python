```python
# Importing the necessary libraries
import pandas as pd
import numpy as np
# Data visualization library
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Reading the dataset in our case it's a csv file
df=pd.read_csv('/content/Life Expectancy Data.csv')
df
```

```python
# head() returns the first record within the dataset
df.head(8)
```

```python
# tail function
# Returns the last record within the data
df.tail(8)

```

```python
# Sanity check

# 3. DATA SANITY CHECK
# a) Dataset Summary Informatiion -> info()
# The Total Number of Rows and Columns
# The Datatype Related to All the Columns
df.info()
```

```python

# b) Missing Records
# Where the Dataset is Empty : True
# Where the Dataset is not Empty: False
# Therefore we sum up all the Columns that Return True, To Get the Total
df.isnull().sum()
```

```python
# we can get the percentage of empty records within a column
# if the column has more than 50% that are empty, then it is dropped
# syntax
# (total empties/total columns)*100
# df.isnull().sum()
(df.isnull().sum()/ df.shape[0]) * 100
(df.isnull().sum()/ df.shape[0]) * 100
percentage = (df.isnull().sum()/ df.shape[0]) * 100
percentage_df = pd.DataFrame(percentage, columns = ["Percentage of Empties"])
percentage_df
```

```python
# Duplicated records
df.duplicated().sum()
```

```python
# 5 Explatory data analysis
# Statistical data analysis
df.describe().transpose()
```

```python
df.describe(include= 'object')
```

```python

# DISTRIBUTION ANALYSIS
# It is only Done on Numerical Columns
# Get the Distribution of Values within the Numerical Columns
# We visualize a Numerical Column on a distplot()
sns.histplot(data=df, x='Life expectancy ', kde=True)
plt.title('Distribution of Life Expectancy')
plt.xlabel('Life Expectancy')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python

# FOR LOOP WITH DISTRIBUTION OF ALL NUMERICAL COLUMNS
df.select_dtypes(include = 'number').columns
for index in df.select_dtypes(include = 'number').columns:
  sns.histplot(data=df, x=df[index], kde=True)
  plt.title(f'Distribution of {index}')
  plt.show()
```
